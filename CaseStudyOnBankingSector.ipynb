```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Age Group': ['18-25', '26-35', '36-45', '46-55', '56-65'],
    'Total Loans': [500, 700, 800, 600, 400],
    'Bad Loans': [50, 60, 40, 30, 20],
    'Good Loans': [450, 640, 760, 570, 380]
}

df = pd.DataFrame(data)

df['% Bad Loans'] = (df['Bad Loans'] / df['Total Loans']) * 100
df['% Good Loans'] = (df['Good Loans'] / df['Total Loans']) * 100
display(df)

max_values = {
    'Age Group': 'Max',
    'Total Loans': max(df['Total Loans']),
    'Bad Loans': max(df['Bad Loans']),
    'Good Loans': max(df['Good Loans']),
    '% Bad Loans': max(df['% Bad Loans']),
    '% Good Loans': max(df['% Good Loans'])
}

df_max = df.replace(max_values, inplace=False)

plt.figure(figsize=(10, 6))
plt.bar(df_max['Age Group'], df_max['Total Loans'], color='skyblue')
plt.title('Total Loans by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Loans')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
